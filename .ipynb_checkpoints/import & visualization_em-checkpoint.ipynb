Time distributions (Hour, Weekend, Month) w/ Seaborn

```python
# Hour distribution
sns.distplot(train['Hour'], hist=True, kde=False, color = 'blue', hist_kws={'edgecolor':'black'}, bins=24)
plt.title('Histogram - Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Record Count')
plt.show()

# Weekend vs. Weekday distribution
sns.distplot(train['Weekend'], hist=True, kde=False, color = 'blue', hist_kws={'edgecolor':'black'}, bins=2)
plt.title('Histogram - Weekend')
plt.xlabel('Weekday = 0 | Weekend = 1')
plt.ylabel('Record Count')
plt.show()

# Month distribution
sns.distplot(train['Month'], hist=True, kde=False, color = 'blue', hist_kws={'edgecolor':'black'}, bins=12)
plt.title('Histogram - Month')
plt.xlabel('Month')
plt.ylabel('Record Count')
plt.show()
```

Records per City

```python
grph = sns.countplot(x='Hour', data=train, hue='City', dodge=True)
grph.set_title("Hour Count Distribution by Week and Weekend Days", fontsize=15)
grph.set_ylabel("Count",fontsize= 12)
grph.set_xlabel("Hours of Day", fontsize=12)
plt.show()
```

## Feature Selection
### Objectives:
- Check correlations
- PCA

```python
# grouping metrics for quicker analysis
TotalTimeStopped = ['TotalTimeStopped_p20','TotalTimeStopped_p50', 'TotalTimeStopped_p80']
TimeFromFirstStop = ['TimeFromFirstStop_p20','TimeFromFirstStop_p50','TimeFromFirstStop_p80']
DistanceToFirstStop = ['DistanceToFirstStop_p20','DistanceToFirstStop_p50','DistanceToFirstStop_p80']
```

Correlation matrix heatmap
- Takeaways:
    - high-correlation between each percentile for time/distance from first stop (as you would expect)
    - high-correlation between 20th percentile for DistanceToFirstStop and TotalTimeStopped

```python
# Checking correlations for metric percentiles
plt.figure(figsize=(15,10))
plt.title('Metric Correlations - Percentiles', fontsize=22)
sns.heatmap(train[TotalTimeStopped + 
                     TimeFromFirstStop + 
                     DistanceToFirstStop].astype(float).corr(),
            vmax=1.0,  annot=True)
plt.show()
```

### Principal Component Analysis

There appear to be some outliers at the upper end of Time Stopped and Distance to First Stop -> need to further check normality

```python
#checking for outliers when considering scaling algorithm
plt.scatter(train['DistanceToFirstStop_p20'], train['TotalTimeStopped_p20'], alpha=0.7)
plt.xlabel('DistanceToFirstStop_p20')
plt.ylabel('TotalTimeStopped_p20')
plt.title('Metric Scatterplots')
plt.show()
```

Further evaluation of normality with QQ Plots

```python
import statsmodels.api as sm
import pylab

sm.qqplot(train['DistanceToFirstStop_p20'],loc=6.6,scale=28.0, line='s')
sm.qqplot(train['DistanceToFirstStop_p50'],loc=28.3,scale=71.7)
sm.qqplot(train['DistanceToFirstStop_p80'],loc=81.9,scale=152.7)
pylab.show()
```

Feature scaling for PCA

```python
# using RobustScaler to scale metrics because of outliers
from sklearn.preprocessing import RobustScaler

cols = TotalTimeStopped + TimeFromFirstStop + DistanceToFirstStop

train_scale = RobustScaler().fit(train[cols])
train_scale.transform(train[cols])
```

```python
train[train['TotalTimeStopped_p80']!=0].head(5)
```

```python
train[cols].columns

train[cols].rename(columns={'TotalTimeStopped_p20':'TotalTimeStopped_p20_s', 'TotalTimeStopped_p50':'TotalTimeStopped_p50_s',
                            'TotalTimeStopped_p80':'TotalTimeStopped_p80_s', 'TimeFromFirstStop_p20':'TimeFromFirstStop_p20_s',
                            'TimeFromFirstStop_p50':'TimeFromFirstStop_p50_s', 'TimeFromFirstStop_p80':'TimeFromFirstStop_p80_s',
                            'DistanceToFirstStop_p20':'DistanceToFirstStop_p20_s', 'DistanceToFirstStop_p50':'DistanceToFirstStop_p50_s',
                            'DistanceToFirstStop_p80':'DistanceToFirstStop_p80_s'})
```

```python
# going to run PCA using the scaled metric columns
from sklearn.decomposition import PCA


pca = PCA(n_components=3, random_state=5)

PCs = pca.fit_transform(train[cols])

principalDf = pd.DataFrame(PCs)

s = 'Target_PCA'

principalDf.rename(columns = lambda x: str(s) + str(x), inplace = True)

train = pd.concat([train, principalDf], axis=1)

train.head(50)
```
