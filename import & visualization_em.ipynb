```python
# 857,409 rows, 28 columns
train.shape
```

```python
# counting missing values by column
# DataFrame.isna() and DataFrame.isnull() are the exact same 
train.isna().sum()

#takeaway: missing values only occur as Street Names (mostly as Entry Street Names)
```

```python
train_na = train[train['EntryStreetName'].isna()]
train_na.head(50)
```

```python
train.nunique()
#Takeaways: 
    # 2,539 intersections in the dataset
    # 4 cities confirmed 
    # only 9 months included

train.groupby('City').count()

# entries by city:
    # Atlanta: 153,363 rows
    # Boston:  182,050 rows
    # Chicago: 133,674 rows
    # Philly:  388,322 rows
```

```python
# more Library imports
import scipy as sp
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# time distributions (Hour, Weekend, Month) w/ seaborn

# Hour distribution
sns.distplot(train['Hour'], hist=True, kde=False, color = 'blue', hist_kws={'edgecolor':'black'}, bins=24)
plt.title('Histogram - Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Record Count')
plt.show()
# Weekend vs. Weekday distribution
sns.distplot(train['Weekend'], hist=True, kde=False, color = 'blue', hist_kws={'edgecolor':'black'}, bins=2)
plt.title('Histogram - Weekend')
plt.xlabel('Weekday = 0 | Weekend = 1')
plt.ylabel('Record Count')
plt.show()
# Month distribution
sns.distplot(train['Month'], hist=True, kde=False, color = 'blue', hist_kws={'edgecolor':'black'}, bins=12)
plt.title('Histogram - Month')
plt.xlabel('Month')
plt.ylabel('Record Count')
plt.show()
```
